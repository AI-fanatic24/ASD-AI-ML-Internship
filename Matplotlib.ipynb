```python
import matplotlib.pyplot as plt
import numpy as np
```

1. Plot a simple line graph using the values: x = [1, 2, 3, 4], y = [10, 20, 25, 30]

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.plot(x,y)
plt.show()
```

2. Add a title and axis labels to the line graph in Question 1.

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.title("LINE GRAPH")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.show()
```

3. Plot a red dotted line for: x = [0, 1, 2, 3], y = [3, 6, 1, 8]

```python
x=[0,1,2,3]
y=[3,6,1,8]
plt.plot(x,y,':r')
plt.show()
```

4. Plot two lines in the same graph: x = [1, 2, 3, 4], y1 = [10, 20, 25, 30], y2 = [5, 15, 20, 25]

```python
x=[1,2,3,4]
y1=[10,20,25,30]
y2=[5,15,20,25]
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

5. Add a legend to the above plot to differentiate the two lines.

```python
x=[1,2,3,4]
y1=[10,20,25,30]
y2=[5,15,20,25]
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(['line1','line2'])
plt.show()
```

11. Create a subplot with two plots: one line plot and one bar chart.

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.subplot(1,2,1)
plt.plot(x,y)

fruits=['Apple','Banana','Mango']
quantity=[10,20,15]
plt.subplot(1,2,2)
plt.bar(fruits,quantity)

plt.show()
```

12. Change the figure size to 8x5 while plotting a simple graph.

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.figure(figsize=(8,5))
plt.plot(x,y)
plt.show()
```

13. Plot a graph with customized line style, marker, and color and save a plot as a PNG file using plt.savefig()

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.plot(x,y,'s--r')
plt.savefig('plot.png') #Must be called before plt.show() otherwise plot will be blank
plt.show()
```

14. Add grid lines to any plot you create.

15. Plot a line graph and annotate the highest point on it.

```python
x=[1,2,3,4]
y=[10,20,25,15]
plt.annotate('PEAK',xy=(3,25))
plt.plot(x,y)
plt.grid()
plt.show()
```

16. Create a bar chart and add value labels on top of each bar.

```python
fruits=['Apple','Banana','Mango']
quantity=[10,20,15]
bars=plt.bar(fruits,quantity)
plt.bar_label(bars,padding=2)
plt.show()
```

17. Create a plot with x-axis rotated 45 degrees using tick labels: Months = ['Jan', 'Feb', 'Mar', 'Apr'], Sales = [250, 300, 150, 400]

```python
months=['Jan','Feb','Mar','Apr']
sales=[250,300,150,400]
plt.xticks(rotation=45)
plt.plot(months,sales)
plt.show()
```

18. Draw three subplots in a single row, each with a different plot (line, bar, pie).

```python
months=['Jan','Feb','Mar','Apr']
sales=[250,300,150,400]
plt.subplot(1,3,1)
plt.plot(months,sales)

fruits=['Apple','Banana','Mango']
quantity=[10,20,15]
plt.subplot(1,3,2)
plt.xticks(rotation=45)
plt.bar(fruits,quantity)

subjects=['Math','Science','English']
values=[30,40,30]
plt.subplot(1,3,3)
plt.pie(values, labels=subjects)

plt.tight_layout()

plt.show()
```

19. Use different colors for each bar in a bar chart using the color parameter.
